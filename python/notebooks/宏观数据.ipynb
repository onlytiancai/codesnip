```python
import pylab as pl
import pandas as pd
from jqdata import finance
from jqdata import macro
from datetime import datetime, timedelta
```

```python
start_date = datetime.now()-timedelta(days=365*3)
end_date = datetime.now()
```

```python
# 防止图表 MathTextWarning
import matplotlib as mpl

mpl.rcParams.update({
    'text.usetex': False,
    'font.family': 'stixgeneral',
    'mathtext.fontset': 'stix',
})
```

```python
def show_data(name, date_field='stat_month', date_format='%Y-%m-01'):
    q = query(name
        ).filter(getattr(name, date_field)>=start_date.strftime(date_format))
    df = macro.run_query(q)
    df.set_index(pd.to_datetime(df[date_field]), inplace=True)
    df.plot(figsize=(16,6))
    pl.legend(loc='lower left',ncol= 5)    
    pl.yscale("log")
    pl.grid()
```
